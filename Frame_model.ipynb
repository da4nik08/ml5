```python
class Airport:
    def __init__(self, name, location, type_, flights_per_day):
        self.name = name
        self.location = location
        self.type = type_
        self.flights_per_day = flights_per_day
```

```python
class Aircraft:
    def __init__(self, model, airline):
        self.model = model
        self.airline = airline

class PassengerAircraft(Aircraft):
    def __init__(self, model, airline, max_passengers):
        super().__init__(model, airline)
        self.max_passengers = max_passengers

class CargoAircraft(Aircraft):
    def __init__(self, model, airline, max_cargo_weight):
        super().__init__(model, airline)
        self.max_cargo_weight = max_cargo_weight

class Drone(Aircraft):
    def __init__(self, model, manufacturer):
        super().__init__(model, manufacturer)

class Helicopter(Aircraft):
    def __init__(self, model, manufacturer):
        super().__init__(model, manufacturer)
```

```python
class AirRoute:
    def __init__(self, number, start_airport, end_airport, distance, travel_time):
        self.number = number
        self.start_airport = start_airport
        self.end_airport = end_airport
        self.distance = distance
        self.travel_time = travel_time
```

```python
class DispatchService:
    def __init__(self, working_zones, checkpoints, flight_management):
        self.working_zones = working_zones
        self.checkpoints = checkpoints
        self.flight_management = flight_management
```

# Recomend 

```python
# Функція для знаходження найближчого аеропорту
def find_nearest_airport(location, *airports):
    nearest_distance = float('inf')
    nearest_airport = None

    for airport in airports:
        distance = calculate_distance(location, airport.location)  # Функція для розрахунку відстані між точками
        if distance < nearest_distance:
            nearest_distance = distance
            nearest_airport = airport

    return nearest_airport
```

```python
# Функція для розрахунку часу подорожі між аеропортами
def calculate_travel_time(start_airport, end_airport, speed):
    distance = calculate_distance(start_airport.location, end_airport.location)
    time = distance / speed  # Припустимо, що швидкість визначає час подорожі
    return time
```

```python
# Функція для вибору оптимального повітряного засобу для маршруту
def choose_optimal_aircraft(route, *aircrafts):
    for aircraft in aircrafts:
        if isinstance(aircraft, PassengerAircraft) and route.distance <= 500:
            return aircraft
        elif isinstance(aircraft, CargoAircraft) and route.distance > 500:
            return aircraft
    return None
```

```python
def calculate_distance(point1, point2):
    _, x1, y1 = point1
    _, x2, y2 = point2
    return math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
```

```python
nearest = find_nearest_airport(["Paris", 25, 25], airport1, airport2)
print(f"The nearest airport to Paris is {nearest.name}")

travel_time = calculate_travel_time(airport1, airport2, 800)  # Припустимо швидкість 800 км/год
print(f"Estimated travel time between {airport1.name} and {airport2.name}: {travel_time} hours")

optimal_aircraft = choose_optimal_aircraft(route1, passenger_plane, cargo_plane, drone)
if optimal_aircraft:
    print(f"The optimal aircraft for route {route1.number} is {optimal_aircraft.model}")
else:
    print("No suitable aircraft found for this route")
```
