```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.DiGraph()
```

# First

```python
concepts = ["Літальний апарат", "Повітряний порт", "Диспетчерська служба", "Маршрут", "Контрольний пункт",
            "Грузовий літак", "Гелікоптер", "Дрон", "Пасажирський літак"]
G.add_nodes_from(concepts)
```

```python
relations = [
    ("Літальний апарат", "Повітряний порт", {'relation': 'літає до'}),
    ("Літальний апарат", "Диспетчерська служба", {'relation': 'повідомляє про'}),
    ("Літальний апарат", "Маршрут", {'relation': 'слідує за'}),
    ("Літальний апарат", "Контрольний пункт", {'relation': 'перевіряє'}),
    ("Грузовий літак", "Літальний апарат", {'relation': 'це'}),
    ("Гелікоптер", "Літальний апарат", {'relation': 'це'}),
    ("Дрон", "Літальний апарат", {'relation': 'це'}),
    ("Пасажирський літак", "Літальний апарат", {'relation': 'це'})
]
```

```python
G.add_edges_from(relations, label='relation')
```

# Second

```python
new_concepts = ["Пасажирський термінал", "Контроль безпеки", "Багаж", "Авіакомпанії", "Служба прийому та відправлення"]
G.add_nodes_from(new_concepts)
```

```python
new_relations = [
    ("Літальний апарат", "Пасажирський термінал", {'relation': 'призначений для'}),
    ("Літальний апарат", "Контроль безпеки", {'relation': 'проходить через'}),
    ("Літальний апарат", "Багаж", {'relation': 'перевозить'}),
    ("Літальний апарат", "Авіакомпанії", {'relation': 'належить до'}),
    ("Літальний апарат", "Служба прийому та відправлення", {'relation': 'співпрацює з'})
]
```

```python
G.add_edges_from(new_relations, label='relation')
```

# Third

```python
additional_concepts = ["Пасажир", "Груз"]
G.add_nodes_from(additional_concepts)
```

```python
additional_relations = [
    ("Пасажирський літак", "Пасажир", {'relation': 'перевозить'}),
    ("Гелікоптер", "Пасажир", {'relation': 'перевозить'}),
    ("Гелікоптер", "Груз", {'relation': 'перевозить'}),
    ("Дрон", "Груз", {'relation': 'перевозить'}),
    ("Грузовий літак", "Груз", {'relation': 'перевозить'})
]
```

```python
G.add_edges_from(additional_relations, label='relation')
```

# Fourth

```python
additional_concepts_2 = ["Вертикальний рух", "Горизонтальний рух", "Міський рух", "Регіональний рух", "Міжнародний рух"]
G.add_nodes_from(additional_concepts_2)
```

```python
additional_relations_2 = [
    ("Гелікоптер", "Вертикальний рух", {'relation': 'має'}),
    ("Пасажирський літак", "Горизонтальний рух", {'relation': 'має'}),
    ("Пасажирський літак", "Міжнародний рух", {'relation': 'має'}),
    ("Пасажирський літак", "Регіональний рух", {'relation': 'має'}),
    ("Грузовий літак", "Міжнародний рух", {'relation': 'має'}),
    ("Грузовий літак", "Регіональний рух", {'relation': 'має'}),
    ("Пасажирський літак", "Міжнародний рух", {'relation': 'має'}),
    ("Гелікоптер", "Міський рух", {'relation': 'має'}),
    ("Гелікоптер", "Регіональний рух", {'relation': 'має'})
]
```

```python
G.add_edges_from(additional_relations_2, label='relation')
```

# Fifth

```python
additional_concepts_3 = ["Служба керування польотами", "Наземні служби управління трафіком", "Контроль висоти та маршрутів"]
G.add_nodes_from(additional_concepts_3)
```

```python
additional_relations_3 = [
    ("Диспетчерська служба", "Служба керування польотами", {'relation': 'включає'}),
    ("Диспетчерська служба", "Наземні служби управління трафіком", {'relation': 'включає'}),
    ("Диспетчерська служба", "Контроль висоти та маршрутів", {'relation': 'включає'})
]
```

```python
G.add_edges_from(additional_relations_3, label='relation')
```

# Sixth

```python
additional_concepts_4 = ["Автопілоти на основі ШІ", "Системи безпеки"]
G.add_nodes_from(additional_concepts_4)
```

```python
additional_relations_4 = [
    ("Пасажирський літак", "Автопілоти на основі ШІ", {'relation': 'використовує'}),
    ("Пасажирський літак", "Системи безпеки", {'relation': 'оснащений'}),
    ("Грузовий літак", "Автопілоти на основі ШІ", {'relation': 'використовує'}),
    ("Грузовий літак", "Системи безпеки", {'relation': 'оснащений'}),
    ("Гелікоптер", "Системи безпеки", {'relation': 'оснащений'}),
    ("Дрон", "Автопілоти на основі ШІ", {'relation': 'використовує'}),
]
```

```python
G.add_edges_from(additional_relations_4, label='relation')
```

# Conclusion

```python
list_of_relations = relations + new_relations + additional_relations + additional_relations_2 + additional_relations_3 + additional_relations_4
count = 0 
for relation in list_of_relations:
    source, target, data = relation
    print(f"Триплет: {source} - {data['relation']} -> {target}")
    count += 1

print(count)
```

```python
pos = nx.spring_layout(G, k=1.0)
nx.draw(G, pos, with_labels=True, node_size=1000, node_color='skyblue', font_weight='bold', font_size=5, arrows=True, width=2.0)
edge_labels = {(i, j): G.get_edge_data(i, j)['relation'] for i, j in G.edges()}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
plt.show()
```
