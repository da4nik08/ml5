```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.DiGraph()
```

```python
concepts = ["Літак", "Повітряний порт", "Диспетчерська служба", "Маршрут", "Контрольний пункт"]
G.add_nodes_from(concepts)
```

```python
relations = [
    ("Літак", "Повітряний порт", {'relation': 'літає до'}),
    ("Літак", "Диспетчерська служба", {'relation': 'повідомляє про'}),
    ("Літак", "Маршрут", {'relation': 'слідує за'}),
    ("Літак", "Контрольний пункт", {'relation': 'перевіряє'})
]
```

```python
G.add_edges_from(relations, label='relation')
```

```python
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=1500, node_color='skyblue', font_weight='bold', font_size=10, arrows=True)
edge_labels = {(i, j): G.get_edge_data(i, j)['relation'] for i, j in G.edges()}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
plt.show()
```
